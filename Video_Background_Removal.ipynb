```
import skvideo.io  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import plotly.express as px
from sklearn.cluster import KMeans
```

```
video = skvideo.io.vread("/content/videoplayback.mp4")  
print(video.shape)
```

```
#Create Background Tensor
background = video[0].copy()
background.shape
```

```
plt.imshow(background)
plt.show()
```

```
n_frames = 10
frames_idx = []
for i in range(0, n_frames):
  frames_idx.append(random.randint(0, video.shape[0]))
  print(random.randint(0, video.shape[0]))
```

```
colors = []
n = 0
for y in range(0, video.shape[1]):
  km = KMeans(n_clusters = 2)
  km = km.fit(colors)
  for z in frames_idx:
    colors.append(video[z][0][0])
    print(n)
    n = n +1
```

```
pixel = pd.DataFrame(colors, columns=["R","G","B"])
```

```
#test
km = KMeans(n_clusters = 2)
km = km.fit(colors)
```

```
km = KMeans(n_clusters = 2)
km = km.fit(pixel)
```
